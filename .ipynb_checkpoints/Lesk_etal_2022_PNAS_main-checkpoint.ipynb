```python
### load energy mix over time and impute final to primary energy

def getFeToPe(rcp):
    #load energy mix
    eM = pd.read_csv(netsetDir+'energy-mix-over-time_'+rcp+'.csv')
    eM = eM.set_index('year')
    #load IEA powerstation inflows and outflows 2014-2018,from https://www.iea.org/sankey/#?c=IEA%20Total&s=Balance
    psInout = pd.read_excel(edgeDir+'pe_to_fe.xlsx').set_index('year')

    #impute PE:FE for non-renewables as average last 5 years ratio inflow to outflows
    feToPe = psInout.out_to_in.mean()

    #estimate overall PE:FE timeseries by weighting by non-renewable share in eMix (assuming renewable PE:FE=1)
    feToPe_t = feToPe*(1-eM[['geoth.','PV','CSP','wind']].sum(axis=1)) + eM[['geoth.','PV','CSP','wind']].sum(axis=1)

    return np.array([feToPe_t.at[t] for t in edge_yrs])
```

```python
### load edge output and declare global mean function:
edgedat = pd.read_csv(edgeDir + 'EDGE_buildings_EDGE_EUR_ETP_v2.csv')

def cddMean(rcp):
    if rcp == 'rcp26':
        cddglob = edgedat[(edgedat.variable=='CDD') & (edgedat.scenario=='SSP2_rcp2p6')&(edgedat.region=='GLO')]#edgedat[(edgedat.variable=='CDD') & (edgedat.scenario=='SSP2_rcp2p6')].groupby(['period']).mean()
    elif rcp == 'rcp45':
        cddglob = edgedat[(edgedat.variable=='CDD') & (edgedat.scenario=='SSP2_rcp4p5')&(edgedat.region=='GLO')]#edgedat[(edgedat.variable=='CDD') & (edgedat.scenario=='SSP2_rcp2p6')].groupby(['period']).mean()
    elif rcp == 'rcp19':
        cddglob = edgedat[(edgedat.variable=='CDD') & (edgedat.scenario=='SSP2_rcp1p9')&(edgedat.region=='GLO')]#edgedat[(edgedat.variable=='CDD') & (edgedat.scenario=='SSP2_rcp2p6')].groupby(['period']).mean()

    cddmean = cddglob[cddglob.period >= start_year]
    cddmean = cddmean.set_index('period')
    return cddmean
```

### sectoral projections

```python
### modeling sea dyke
km_to_m = 10**3
#year adaptations are built (assumed to be midpoint of transition)
t_adapt = 2060
#percent intital adaptation
adapt_initial_frac = 0.5
#thickness of outer dyke revetement (m)
T = 1
#slopes inverse of dike (1/3 and 1/4)
m1 = 3
m2 = 4
#ratio of dyke height to foundation thickness
B = 0.25

#function to compute sea wall concrete volume
def seaDikeConcreteVol(hopt,L,SLR,S1,wc):
    
    #actual height of seawall (optimal protection + wave runup + inital under-adaptation)
    H=hopt+wc+adapt_initial_frac*S1
    #volume of concrete in dyke per unit L (revetement + foundation)
    Vc= ((m1+m2)*H*T-0.5*(m1+m2)*T**2+(m1+m2)*T) + B*H*(m1+m2+5)*H
    return Vc*L*km_to_m
    
```

```python
### coastal adaptation energy accounting:
#embodied carbon and energy from Hammond et al 2008
energy_emb_concrete = 0.95 #MJ/kg concrete, standard 1:2:4
energy_emb_steel = 24.4 #MJ/kg general typical 42.3% recycled

#conversions
rho_concrete = 2400 #kg/m3
MJ_to_EJ = 10**-12
t_to_kg = 1000
energy_emb_concrete = energy_emb_concrete * rho_concrete # -> MJ/m3 concrete
steel_per_concrete = 150 #kg steel/m3 concrete (ecoinvent ref https://db.ecoinvent.org/reports/07_BuildingProducts.pdf?area=463ee7e58cbf8)
energy_emb_steel = energy_emb_steel * steel_per_concrete  # -> MJ/m3 concrete

#function to embedded energy in sea dike over each coastline segment
def sdEnergy(rcp,p):
    ### load full coastal dataset
    fuldat = pd.read_csv(ciamOutDir+'ciam-output-'+rcp+'_'+p+'_full.csv')
    fuldat = fuldat.set_index('Unnamed: 0')

    fuldat['seaDikeEnergy'] = np.nan
    
    #loop over segments
    for i in fuldat.index:

        if fuldat.loc[i].optA != 'fullProtect': continue
        else:
            hopt = fuldat.loc[i].value
            SLR = fuldat.loc[i].slr
            L = fuldat.loc[i].L
            S1 = fuldat.loc[i].s1
            wc = fuldat.loc[i].waveclim
            fuldat.at[i,'seaDikeEnergy'] = seaDikeConcreteVol(hopt,L,SLR,S1,wc)*(energy_emb_steel+energy_emb_concrete)


    return fuldat[fuldat.optA == 'fullProtect'].seaDikeEnergy

```

```python
### model sea dike concrete emissions factors (process vs. energy emissions) over the transition

### factors and conversions
#emissions intensities of fuels (non time varying)
EFcoal = EFs['coal']*1e3
EFoil = EFs['oil']*1e3 
EFgas = EFs['gas']*1e3

#embodied carbon and energy from Hammond et al 2008
carb_emb_concrete = 0.035 #kgC/kg concrete, standard 1:2:4 
energy_emb_concrete = 0.95 #MJ/kg 
carb_emb_steel = 0.482 #kgC/kg steel, general typical 42.3% recycled
energy_emb_steel = 24.4 #MJ/kg

#conversions
rho_concrete = 2400 #kg/m3
c_to_co2 = 44/12 
MJ_to_TJ = 10**-6 
MJ_to_EJ = 10**-12
t_to_kg = 1000
carb_emb_concrete = carb_emb_concrete * c_to_co2 * rho_concrete # -> kg co2/m3 concrete
energy_emb_concrete = energy_emb_concrete * rho_concrete # -> MJ/m3 concrete

steel_per_concrete = 150 #kg steel/m3 concrete (ecoinvent ref https://db.ecoinvent.org/reports/07_BuildingProducts.pdf?area=463ee7e58cbf8)
carb_emb_steel = carb_emb_steel * c_to_co2 * steel_per_concrete # -> kg co2/m3 concrete
energy_emb_steel = energy_emb_steel * steel_per_concrete  # -> MJ/m3 concrete

tinit = 2010 #year to assume starting energy mix for updating energy emissions in material EFs

### function to re-estimate total EF (kgCO2 / m3) concrete through the transition
### returns sepeartely: concrete_chem, steel_chem, energy
def getConcreteEF(t):
    
    eMix = pd.read_csv(netsetDir+'energy-mix-over-time_'+rcp+'.csv')
    eMix = eMix.set_index('year')
    
    #update global energy EF based on energy mix at time t
    energyEFmean = (EFcoal*eMix.loc[t].coal + EFoil*eMix.loc[t].oil + EFgas*eMix.loc[t].gas)
    energyEFmean_init = (EFcoal*eMix.loc[tinit].coal + EFoil*eMix.loc[tinit].oil + EFgas*eMix.loc[tinit].gas)

    #subtract embodied-energy emissions from carbon embodied to isolate chemical component
    carb_emb_concrete_chem = carb_emb_concrete - energyEFmean_init*energy_emb_concrete*MJ_to_TJ
    carb_emb_steel_chem = carb_emb_steel - energyEFmean_init*energy_emb_steel*MJ_to_TJ
    
    #add energy emissions back in with energy mix at time t
    return carb_emb_concrete_chem, carb_emb_steel_chem, energyEFmean*energy_emb_concrete*MJ_to_TJ, energyEFmean*energy_emb_steel*MJ_to_TJ
    
time = np.arange(2020,2101,10)

#re-estimate emissions factors over time
ef_vect = [getConcreteEF(t) for t in time]
    
conc_chem = [a[0] for a in ef_vect]
steel_chem = [a[1] for a in ef_vect]
conc_energy = [a[2] for a in ef_vect]
steel_energy = [a[3] for a in ef_vect]
```

```python
### partition change drivers between 2 scenarios: retreat

#redefine functions to handle sensitivity test
#modified concrete EF function takes in a specified RCP scenario
def getConcreteEF_s(rcp,t):
    eMix = pd.read_csv(netsetDir+'energy-mix-over-time_'+rcp+'.csv')
    eMix = eMix.set_index('year')
    #update global energy EF based on energy mix at time t
    energyEFmean = (EFcoal*eMix.loc[t].coal + EFoil*eMix.loc[t].oil + EFgas*eMix.loc[t].gas)
    energyEFmean_init = (EFcoal*eMix.loc[tinit].coal + EFoil*eMix.loc[tinit].oil + EFgas*eMix.loc[tinit].gas)
    #subtract embodied-energy emissions from carbon embodied to isolate chemical component
    carb_emb_concrete_chem = carb_emb_concrete - energyEFmean_init*energy_emb_concrete*MJ_to_TJ
    carb_emb_steel_chem = carb_emb_steel - energyEFmean_init*energy_emb_steel*MJ_to_TJ
    #add energy emissions back in with energy mix at time t
    return carb_emb_concrete_chem, carb_emb_steel_chem, energyEFmean*energy_emb_concrete*MJ_to_TJ, energyEFmean*energy_emb_steel*MJ_to_TJ

#modified sdEms handles seperate scenario RCP for emissions intensity vs. coastal projection
def sdEms_s(rcpN,rcpI,p):
    ### load full coastal dataset
    fuldat = pd.read_csv(ciamOutDir+'ciam-output-'+rcpN+'_'+p+'_full.csv')
    fuldat = fuldat.set_index('Unnamed: 0')
    fuldat['seaDikeEmissions'] = np.nan
    for i in fuldat.index:
        if fuldat.loc[i].optA != 'fullProtect': continue
        else:
            hopt = fuldat.loc[i].value
            SLR = fuldat.loc[i].slr
            L = fuldat.loc[i].L
            S1 = fuldat.loc[i].s1
            wc = fuldat.loc[i].waveclim
            fuldat.at[i,'seaDikeEmissions'] = seaDikeConcreteVol(hopt,L,SLR,S1,wc)*sum(getConcreteEF_s(rcpI,t_adapt))
    return fuldat[fuldat.optA == 'fullProtect'].seaDikeEmissions.sum()/10**12

### compute energy and emissions for both rcps
#protectionEnergy = sdEnergy('rc___','p50').sum()*MJ_to_EJ
#retreatEnergy = energy_dwelling*area_dwelling*retreat_pop/people_per_dwelling*GJ_to_EJ

ePm = sdEms_s('rcp26','rcp19','p50')
eP19 = sdEms_s('rcp19','rcp19','p50')

ePDelI19 = (eP26-ePm)#/(eR46-eR26)
ePDelN19 = (ePm-eP19)

print(ePDelI19/(eP26-eP19))
print(ePDelN19/(eP26-eP19))
print(eP45)
print(ePm)
```
