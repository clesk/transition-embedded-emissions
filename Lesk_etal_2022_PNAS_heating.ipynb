```python
### aggregate carriers to emissions factor category, per scenario

#electricity
elecTotal_19 = (elecRH_19_C - elecRH_19_noC) + (elecHP_19_C - elecHP_19_noC) + (elecDH_19_C - elecDH_19_noC) 
elecTotal_26 = (elecRH_26_C - elecRH_26_noC) + (elecHP_26_C - elecHP_26_noC) + (elecDH_26_C - elecDH_26_noC) 
elecTotal_45 = (elecRH_45_C - elecRH_45_noC) + (elecHP_45_C - elecHP_45_noC) + (elecDH_45_C - elecDH_45_noC) 

#biomod
biomodTotal_19 = (biomod_19_C - biomod_19_noC) + (bioDH_19_C - bioDH_19_noC)
biomodTotal_26 = (biomod_26_C - biomod_26_noC) + (bioDH_26_C - bioDH_26_noC)
biomodTotal_45 = (biomod_45_C - biomod_45_noC) + (bioDH_45_C - bioDH_45_noC)

#biotrad
biotrad_19 = (biotrad_19_C - biotrad_19_noC)
biotrad_26 = (biotrad_26_C - biotrad_26_noC)
biotrad_45 = (biotrad_45_C - biotrad_45_noC)

#coal
coal_19 = coal_19_C - coal_19_noC
coal_26 = coal_26_C - coal_26_noC
coal_45 = coal_45_C - coal_45_noC

#oil
oil_19 = oil_19_C - oil_19_noC
oil_26 = oil_26_C - oil_26_noC
oil_45 = oil_45_C - oil_45_noC

#gas
gas_19 = gas_19_C - gas_19_noC
gas_26 = gas_26_C - gas_26_noC
gas_45 = gas_45_C - gas_45_noC
```

```python
### define emissions factors
#kgCO2/TJ
EFs = {'coal':90*10**3,'oil':60*10**3,'gas':50*10**3,'biotrad':15*10**3,'biomod':25*10**3}

EJ_to_TJ = 10**6
kgCO2_to_GtCO2 = 10**-12

#update electricity emissions intensity based on global energy mix
def getEnergyEF(t,scen):
    
    #load energy mix over time
    eMix = pd.read_csv(netsetDir+'energy-mix-over-time_'+scen+'.csv')
    eMix = eMix.set_index('year')
    
    #update global energy EF based on energy mix at time t
    return (EFs['coal']*eMix.loc[t].coal + EFs['oil']*eMix.loc[t].oil + EFs['gas']*eMix.loc[t].gas)

```

```python
### convert final to primary energy for emissions calcs -- these are different than main analysis due to peculiarities of heating
def getFeToPe(rcp):
    #load energy mix
    eM = pd.read_csv(netsetDir+'energy-mix-over-time_'+rcp+'.csv')
    eM = eM.set_index('year')
    #load IEA powerstation inflows and outflows 2014-2018,from https://www.iea.org/sankey/#?c=IEA%20Total&s=Balance
    psInout = pd.read_excel(edgeDir+'pe_to_fe.xlsx').set_index('year')

    #impute PE:FE for non-renewables as average last 5 years ratio inflow to outflows
    feToPe = psInout.out_to_in.mean()

    #estimate overall PE:FE timeseries by weighting by non-renewable share in eMix (assuming renewable PE:FE=1)
    feToPe_t = feToPe*(1-eM[['geoth.','PV','CSP','wind']].sum(axis=1)) + eM[['geoth.','PV','CSP','wind']].sum(axis=1)

    return np.array([feToPe_t.at[t] for t in yr])
```

```python
### calculate emissions (convert EJ to TJ as emissions factors in kgCO2/TJ, then convert kgCO2 to GtCO2 for final units)
start_index = 20

#rcp1.9
ems_elec19ts = elecTotal_19[start_index:].values.reshape(17)*np.array([getEnergyEF(t,'rcp19') for t in yr])*EJ_to_TJ*kgCO2_to_GtCO2/getFeToPe('rcp19')
ems_biomod19ts = biomodTotal_19[start_index:].values.reshape(17)*EFs['biomod']*EJ_to_TJ*kgCO2_to_GtCO2
ems_biotrad19ts = biotrad_19[start_index:].values.reshape(17)*EFs['biotrad']*EJ_to_TJ*kgCO2_to_GtCO2
ems_coal19ts = coal_19[start_index:].values.reshape(17)*EFs['coal']*EJ_to_TJ*kgCO2_to_GtCO2
ems_oil19ts = oil_19[start_index:].values.reshape(17)*EFs['oil']*EJ_to_TJ*kgCO2_to_GtCO2
ems_gas19ts = gas_19[start_index:].values.reshape(17)*EFs['gas']*EJ_to_TJ*kgCO2_to_GtCO2

#rcp2.6
ems_elec26ts = elecTotal_26[start_index:].values.reshape(17)*np.array([getEnergyEF(t,'rcp26') for t in yr])*EJ_to_TJ*kgCO2_to_GtCO2/getFeToPe('rcp26')
ems_biomod26ts = biomodTotal_26[start_index:].values.reshape(17)*EFs['biomod']*EJ_to_TJ*kgCO2_to_GtCO2
ems_biotrad26ts = biotrad_26[start_index:].values.reshape(17)*EFs['biotrad']*EJ_to_TJ*kgCO2_to_GtCO2
ems_coal26ts = coal_26[start_index:].values.reshape(17)*EFs['coal']*EJ_to_TJ*kgCO2_to_GtCO2
ems_oil26ts = oil_26[start_index:].values.reshape(17)*EFs['oil']*EJ_to_TJ*kgCO2_to_GtCO2
ems_gas26ts = gas_26[start_index:].values.reshape(17)*EFs['gas']*EJ_to_TJ*kgCO2_to_GtCO2

#rcp4.5
ems_elec45ts = elecTotal_45[start_index:].values.reshape(17)*np.array([getEnergyEF(t,'rcp45') for t in yr])*EJ_to_TJ*kgCO2_to_GtCO2/getFeToPe('rcp45')
ems_biomod45ts = biomodTotal_45[start_index:].values.reshape(17)*EFs['biomod']*EJ_to_TJ*kgCO2_to_GtCO2
ems_biotrad45ts = biotrad_45[start_index:].values.reshape(17)*EFs['biotrad']*EJ_to_TJ*kgCO2_to_GtCO2
ems_coal45ts = coal_45[start_index:].values.reshape(17)*EFs['coal']*EJ_to_TJ*kgCO2_to_GtCO2
ems_oil45ts = oil_45[start_index:].values.reshape(17)*EFs['oil']*EJ_to_TJ*kgCO2_to_GtCO2
ems_gas45ts = gas_45[start_index:].values.reshape(17)*EFs['gas']*EJ_to_TJ*kgCO2_to_GtCO2
```

```python
plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.size'] = 18
start_index = 20

colors = ['forestgreen','sandybrown','saddlebrown','#7a0177','#ae017e','#dd3497']

axs = []
fig = plt.figure(figsize=(22,13))
spec = gridspec.GridSpec(ncols=5, nrows=3)

axs.append(fig.add_subplot(spec[0, 0:2]))
axs.append(fig.add_subplot(spec[1, 0:2]))
axs.append(fig.add_subplot(spec[2, 0:2]))
axs.append(fig.add_subplot(spec[0, 2:4]))
axs.append(fig.add_subplot(spec[1, 2:4]))
axs.append(fig.add_subplot(spec[2, 2:4]))
axs.append(fig.add_subplot(spec[0, 4]))
axs.append(fig.add_subplot(spec[1, 4]))
axs.append(fig.add_subplot(spec[2, 4]))

### energy timeseries

#a) rapid scenario 1.9
ax = axs[0]
vs = [elecTotal_19,biomodTotal_19,biotrad_19,coal_19,oil_19,gas_19]

for i,v in enumerate(vs):
    ax.plot(yr, v[start_index:],color=colors[i])
    
ax.legend(['Electricity','Mod. biofuels','Trad. biofuels','Coal','Oil','Gas'])
ax.set_title('Energy demand avoided from adaptive heating')
ax.set_ylabel('Primary energy (EJ/yr)')
ax.set_ylim([-1.7,0.2])

#b) gradual scenario 2.6
ax = axs[1]
vs = [elecTotal_26,biomodTotal_26,biotrad_26,coal_26,oil_26,gas_26]

for i,v in enumerate(vs):
    ax.plot(yr, v[start_index:],color=colors[i])

ax.set_ylabel('Primary energy (EJ/yr)')
ax.set_ylim([-1.7,0.2])

#c) delayed scenario 4.5
ax = axs[2]
vs = [elecTotal_45,biomodTotal_45,biotrad_45,coal_45,oil_45,gas_45]

for i,v in enumerate(vs):
    ax.plot(yr, v[start_index:],color=colors[i])
    
ax.set_ylabel('Primary energy (EJ/yr)')
ax.set_ylim([-1.7,0.2])

### emissions timeseries

#d) 1.9
ax = axs[3]
vs = [ems_elec19ts,ems_biomod19ts,ems_biotrad19ts,ems_coal19ts,ems_oil19ts,ems_gas19ts]

for i,v in enumerate(vs):
    ax.plot(yr, v,color=colors[i])
    
ax.set_title('Emissions avoided from adaptive heating')
ax.set_ylabel(r'$GtCO_2$')
ax.set_ylim([-0.035,0.005])

#e) 2.6
ax = axs[4]
vs = [ems_elec26ts,ems_biomod26ts,ems_biotrad26ts,ems_coal26ts,ems_oil26ts,ems_gas26ts]

for i,v in enumerate(vs):
    ax.plot(yr, v,color=colors[i])

ax.set_ylabel(r'$GtCO_2$')
ax.set_ylim([-0.035,0.005])

#f) 4.5
ax = axs[5]
vs = [ems_elec45ts,ems_biomod45ts,ems_biotrad45ts,ems_coal45ts,ems_oil45ts,ems_gas45ts]

for i,v in enumerate(vs):
    ax.plot(yr, v,color=colors[i])

ax.set_ylabel(r'$GtCO_2$')
ax.set_ylim([-0.035,0.005])

### emissions totals

#g) rapid scenario rcp1.9
ax = axs[6]
vs = [ems_elec19ts,ems_biomod19ts,ems_biotrad19ts,ems_coal19ts,ems_oil19ts,ems_gas19ts]

ax.bar([0,1],[sum(sum(vs)),0.2*0.3],color=['orange','coral'])
ax.set_xlim(-0.5,1.5)
ax.set_xticks([0,1])
ax.set_xticklabels(['Heating','Cooling'])
ax.hlines(0,-0.5,1.5,color='gray')
ax.set_ylabel(r'$GtCO_2$')
ax.set_ylim([-2.4,2.1])

ax.set_title('Total, 2020-2100')

#g) gradual rcp2.6
ax = axs[7]
vs = [ems_elec26ts,ems_biomod26ts,ems_biotrad26ts,ems_coal26ts,ems_oil26ts,ems_gas26ts]

ax.bar([0,1],[sum(sum(vs)),0.3],color=['orange','coral'])
ax.set_xlim(-0.5,1.5)
ax.set_xticks([0,1])
ax.set_xticklabels(['Heating','Cooling'])
ax.hlines(0,-0.5,1.5,color='gray')
ax.set_ylabel(r'$GtCO_2$')
ax.set_ylim([-2.4,2.1])

#g) delayed rcp4.5
ax = axs[8]
vs = [ems_elec45ts,ems_biomod45ts,ems_biotrad45ts,ems_coal45ts,ems_oil45ts,ems_gas45ts]

ax.bar([0,1],[sum(sum(vs)),2.0],color=['orange','coral'])
ax.set_xlim(-0.5,1.5)
ax.set_xticks([0,1])
ax.set_xticklabels(['Heating','Cooling'])
ax.hlines(0,-0.5,1.5,color='gray')
ax.set_ylabel(r'$GtCO_2$')
ax.set_ylim([-2.4,2.1])

plt.tight_layout()

abc = 'abcdefghi'
for i,ax in enumerate(axs):
    if i < 8:
        ax.text(0.03, 0.1, abc[i]+')',fontweight='bold',
        horizontalalignment='left',
        verticalalignment='top',
        transform=ax.transAxes,size=25,alpha=0.7)
    else:
        axs[8].text(0.9, 0.1, 'i)',fontweight='bold',
        horizontalalignment='left',
        verticalalignment='top',
        transform=ax.transAxes,size=25,alpha=0.7)
    
#plt.savefig('C:\\Users\\corey\\Documents\\research\\adfb\\manuscript\\pnas\\r1\\figs\\heating.png')
```
